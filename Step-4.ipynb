```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt

try:
    image = imageio.imread('image.png')

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(image)
    plt.title("Original Image")
    plt.axis("off")

    inverted_image = 255 - image 

    plt.subplot(122)
    plt.imshow(inverted_image)
    plt.title("Inverted Image")
    plt.axis("off")

    plt.show()
except FileNotFoundError:
    print("image.png file not found")
```

```python
import numpy as np
import imageio.v2 as imageio
import matplotlib.pyplot as plt

try: 
    image = imageio.imread('image.png')

    plt.figure(figsize=(10, 5))
    plt.title("Original Image")
    plt.imshow(image)
    plt.subplot(121)
    plt.axis("off")

    plt.figure(figsize=(10, 5))
    plt.title("Inverted Image")
    plt.imshow(1 - image)
    plt.subplot(121)
    plt.axis("off")

    plt.show()
except:
    print("image.png file not found") 
```
